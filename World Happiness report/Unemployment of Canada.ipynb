```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the file is named "unemployment.csv" and is in the same directory as the script
file_path = "unemployment.csv"

# Read the CSV file
df = pd.read_csv(file_path)

# Filter the data for the year 2021
df_2021 = df[df['REF_DATE'].str.startswith('2021')]

# Group the data by GEO (province) and calculate the average Employment Rate and Unemployment Rate for each province
province_data = df_2021.groupby('GEO').agg({'Employment rate': 'mean', 'Unemployment rate': 'mean'})

# Plotting the data
fig, axs = plt.subplots(2, 1, figsize=(12, 12))

# Full-time employment graph
province_data['Employment rate'].plot(kind='bar', ax=axs[0], color='b')
axs[0].set_ylabel('Employment Rate')
axs[0].set_title('Employment rate by Province in Canada (2021)')

# Unemployment graph
province_data['Unemployment rate'].plot(kind='bar', ax=axs[1], color='r')
axs[1].set_ylabel('Unemployment rate')
axs[1].set_title('Unemployment by Province in Canada (2021)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the file is named "unemployment.csv" and is in the same directory as the script
file_path = "unemployment.csv"

# Read the CSV file
df = pd.read_csv(file_path)

# Filter the data for the year 2021 and full-time employment up to 1,000,000
df_2021 = df[(df['REF_DATE'].str.startswith('2021')) & (df['Full-time employment'] <= 1000000)]

# Group the data by GEO (province) and calculate the total full-time employment for each province
province_data = df_2021.groupby('GEO').agg({'Full-time employment': 'sum',})

# Plotting the data
fig, axs = plt.subplots(2, 1, figsize=(12, 12))

# Full-time employment graph
province_data['Full-time employment'].plot(kind='bar', ax=axs[0], color='b')
axs[0].set_ylabel('Full-time Employment')
axs[0].set_title('Full-time Employment by Province in Canada (2021, up to 1,000,000)')


plt.tight_layout()
plt.show()
```
