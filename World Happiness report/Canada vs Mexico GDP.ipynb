```python
import pandas as pd
import matplotlib.pyplot as plt

# List of file names (years)
years = ['2021', '2022', '2020', '2019', '2018', '2017']

# List to store individual DataFrames
dataframes = []

# Loop through each year and read the corresponding CSV file
for year in years:
    file_path = f'{year}.csv'  # Construct the file path
    df = pd.read_csv(file_path)  # Read the CSV file
    df['Year'] = int(year)  # Add a 'Year' column to each DataFrame
    dataframes.append(df)  # Append the DataFrame to the list

# Combine all DataFrames into a single DataFrame
df_combined = pd.concat(dataframes, ignore_index=True)

# Check the column names to ensure they match what you expect
print(df_combined.columns)

# Filter the DataFrame for Canada and Mexico
df_canada = df_combined[df_combined['Country name'] == 'Canada']
df_mexico = df_combined[df_combined['Country name'] == 'Mexico']

# Sort the DataFrames by year
df_canada = df_canada.sort_values(by='Year')
df_mexico = df_mexico.sort_values(by='Year')

# Plot the ladder scores over the years
plt.figure(figsize=(10, 6))
plt.plot(df_canada['Year'], df_canada['Ladder score'], marker='o', label='Canada')
plt.plot(df_mexico['Year'], df_mexico['Ladder score'], marker='o', label='Mexico')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Ladder Score')
plt.title('Ladder Scores of Canada and Mexico (2017-2022)')
plt.legend()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the file is named "economy_data.csv" and is in the same directory as the script
file_path = "gdp.csv"

# Read the CSV file with the first column as the index
data = pd.read_csv(file_path, index_col=0)

# Filter the data for Canada and the years 2010 to 2018
canada_data = data.loc['Canada', '2010':'2018']

# Convert data values to float
canada_data = canada_data.astype(float)

# Create a line graph
plt.figure(figsize=(12, 6))
plt.plot(canada_data.index, canada_data.values, marker='o')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP of Canada (2010-2018)')
plt.grid(True)
plt.show()



```
