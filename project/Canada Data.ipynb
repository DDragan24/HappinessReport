```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data ="global_inflation_data.csv"
globalinflationdata = pd.read_csv(data)
globalinflationdata.head()
```

```python
Countrydf = globalinflationdata.loc[globalinflationdata["country_name"]=="Canada"]
Countrydf
```

```python
Countrydf2=Countrydf.set_index("country_name")
Countrydf2=Countrydf2.transpose()
Countrydf2.head()
```

```python
Countrydf2.head()
Countrydf2=Countrydf2.iloc[1:,:]
```

```python
Countrydf2.rename(columns={"country_name":"year"},inplace=True)
Countrydf2.reset_index(inplace=True)
```

```python
Countrydf2["index"]=Countrydf2["index"].astype(int)
Countrydf2.dtypes
```

```python
Countrydfyear=Countrydf2.loc[Countrydf2["index"]>=2013]
Countrydfyear
```

```python
plt.plot(Countrydfyear["index"], Countrydfyear["Canada"])
plt.show()
```
